## Exercițiul 3

Construiți un semnal sinusoidal de frecvență aleasă de voi, de amplitudine unitară și fază nulă. Realizați un grafic care să arăte că eșantionarea lui cu o **frecvență sub-Nyquist** (aleasă, de asemenea, de voi) generează fenomenul de aliere.

În acest scop, generați alte două semnale sinusoidale (cu frecvențe fundamentale _diferite_ față de cea a semnalului inițial) care eșantionate cu frecvența aleasă mai sus produc **aceleași eșantioane** ca semnalul inițial. Ar trebui să obțineți o situație similară cu cea din Figura 2 din PDF.

```python
import numpy as np
import matplotlib.pyplot as plt

f_original=10

f_esantionare=8 # 2*10=20>8
f_alias1=abs(f_original-f_esantionare)
f_alias2=f_original+f_esantionare

t=np.linspace(0,2,500)
semnal_original=np.sin(2*np.pi*f_original*t)
semnal_alias1=np.sin(2*np.pi*f_alias1*t)
semnal_alias2=np.sin(2*np.pi*f_alias2*t)

esantioane=np.arange(int(2*f_esantionare)) / f_esantionare
semnal_esantionat=np.sin(2*np.pi*f_original*esantioane)

plt.figure()
fig, axs=plt.subplots(4, 1)
axs[0].plot(t, semnal_original)
axs[0].stem(esantioane, semnal_esantionat)

axs[1].plot(t, semnal_alias1)
axs[1].stem(esantioane, semnal_esantionat)

axs[2].plot(t, semnal_alias2)
axs[2].stem(esantioane, semnal_esantionat)

axs[3].stem(esantioane, semnal_esantionat)
plt.ylabel('Amplitutide')
plt.xlabel('Timp')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Exercițiul 4

Eșantionați încă o dată semnalele de la exercițiul anterior, de data aceasta alegând o frecvență de eșantionare **mai mare decât frecvența Nyquist**, astfel încât să nu mai obțineți fenomenul de aliere. Reprezentați-le grafic pentru a demonstra că nu mai are loc alierea.

```python
import numpy as np
import matplotlib.pyplot as plt

f_original=10

f_esantionare=40 # 2*10=20>8
f_alias1=abs(f_original-f_esantionare)
f_alias2=f_original+f_esantionare

t=np.linspace(0,0.5,500)
esantioane=np.arange(0, 0.5, 1/f_esantionare)

semnal_original=np.sin(2*np.pi*f_original*t)
semnal_alias1=np.sin(2*np.pi*f_alias1*t)
semnal_alias2=np.sin(2*np.pi*f_alias2*t)

semnal_esantionat=np.sin(2*np.pi*f_original*esantioane)
semnal_alias1_esantionat=np.sin(2*np.pi*f_alias1*esantioane)
semnal_alias2_esantionat=np.sin(2*np.pi*f_alias2*esantioane)

plt.figure()
fig, axs=plt.subplots(3, 1)
axs[0].plot(t, semnal_original)
axs[0].stem(esantioane, semnal_esantionat)

axs[1].plot(t, semnal_alias1)
axs[1].stem(esantioane, semnal_alias1_esantionat)

axs[2].plot(t, semnal_alias2)
axs[2].stem(esantioane, semnal_alias2_esantionat)
plt.ylabel('Amplitutide')
plt.xlabel('Timp')
plt.grid(True)
plt.tight_layout()
plt.show()
```
