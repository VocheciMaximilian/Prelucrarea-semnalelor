## Exercițiul 1

Fie semnalele continue $x(t) = \cos (520 \pi t + \pi/3)$, $y(t) = \cos (280 \pi t - \pi/3)$ și $z(t) = \cos (120 \pi t + \pi/3)$.

- În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu $[0:0.0005:0.03]$ (adică intervalul $[0, 0.03]$, discretizat cu un pas de $0.0005$).

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0, 0.03, 0.0005)
```

- Construiți semnalele $x(t)$, $y(t)$ și $z(t)$ și afișați-le grafic, în câte un subplot.

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0, 0.03, 0.0005)

Xt=np.cos(520*np.pi*t+np.pi/3)
Yt=np.cos(280*np.pi*t-np.pi/3)
Zt=np.cos(120*np.pi*t+np.pi/3)

fig, axs=plt.subplots(3, 1, figsize=(10,8))

axs[0].plot(t,Xt)
axs[0].set_title('X(t)')
axs[0].set_xlabel('Timp')
axs[0].set_ylabel('Ampl')
axs[0].grid(True)

axs[1].plot(t,Yt)
axs[1].set_title('Y(t)')
axs[1].set_xlabel('Timp')
axs[1].set_ylabel('Ampl')
axs[1].grid(True)

axs[2].plot(t,Zt)
axs[2].set_title('Z(t)')
axs[2].set_xlabel('Timp')
axs[2].set_ylabel('Ampl')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs=200
Ts=1/fs
n=np.arange(0, 0.03, Ts)

Xt=np.cos(520*np.pi*n+np.pi/3)
Yt=np.cos(280*np.pi*n-np.pi/3)
Zt=np.cos(120*np.pi*n+np.pi/3)

fig, axs=plt.subplots(3, 1, figsize=(10,8))

axs[0].plot(n,Xt)
axs[0].set_title('X(n)')
axs[0].set_xlabel('Timp')
axs[0].set_ylabel('Ampl')
axs[0].grid(True)

axs[1].plot(n,Yt)
axs[1].set_title('Y(t)')
axs[1].set_xlabel('Timp')
axs[1].set_ylabel('Ampl')
axs[1].grid(True)

axs[2].plot(n,Zt)
axs[2].set_title('Z(t)')
axs[2].set_xlabel('Timp')
axs[2].set_ylabel('Ampl')
axs[2].grid(True)

plt.tight_layout()
plt.show()

#graficele sunt similare datorita fenomenului de aliasing?
```

## Exercițiul 2

Generați semnale descrise mai jos și afișați-le grafic, fiecare într-un plot. Nu aveți voie să folosiți funcțiile predefinite din `scipy.signal`.

- Un semnal **sinusoidal** de frecvență 400 Hz, care să conțină 1600 de eșantioane.

```python
import numpy as np
import matplotlib.pyplot as plt

fs=400
f_esantionare=40000
nr_esantioane=1600
n=np.arange(nr_esantioane)
t=n/f_esantionare

semnal=np.sin(2*np.pi*fs*t)
fig, axs=plt.subplots(1, 1, figsize=(10,8))

axs.plot(t,semnal)
axs.set_xlabel('Timp')
axs.set_ylabel('Ampl')
axs.grid(True)
plt.show()
```

- Un semnal **sinusoidal** de frecvență 800 Hz, care să dureze 3 secunde.

```python
import numpy as np
import matplotlib.pyplot as plt

fs=800
f_esantionare=1800
t=np.arange(0, 3, 1/f_esantionare)
A=1
semnal=A*np.sin(2*np.pi*fs*t)
fig, axs=plt.subplots(1, 1, figsize=(10,8))

axs.plot(t,semnal)
axs.set_xlabel('Timp')
axs.set_ylabel('Ampl')
axs.grid(True)
plt.show()
```

- Un semnal de tip **sawtooth** de frecvență 240 Hz (puteți folosi funcțiile `numpy.floor` sau `numpy.mod`).

```python
import numpy as np
import matplotlib.pyplot as plt

fs=240
f_esantionare=4000
t=np.arange(0, 0.03, 1/f_esantionare)
saw=2*((t*fs)%1)-1
saw*=0.99
plt.plot(t, saw)
plt.show()
```

- Un semnal de tip **square** de frecvență 300 Hz (puteți folosi funcția `numpy.sign`).

```python
import numpy as np
import matplotlib.pyplot as plt

fs=300
f_esantionare=44000
t=np.arange(0, 0.03, 1/f_esantionare )

s_sin=np.sin(2*np.pi*fs*t)
s_squ=np.sign(s_sin)

plt.plot(t, s_squ)
plt.show
```

- Un semnal 2D **la alegerea voastră**. Creați un `numpy.array` de dimensiune 128x128 și inițializați-l folosind o procedură creată de voi.
Utilizați, spre exemplu, funcțiile `numpy.zeros()` și `numpy.ones()`.

```python
import numpy as np
import matplotlib.pyplot as plt
n=128
m=128
img=np.zeros((n, m))
centru=np.floor(n/2)
grosime=30
start=int(centru-np.floor(grosime/2))
end=int(centru+np.floor(grosime/2))
img[:, start:end]=1
img[start:end, :]=1
for i in range(n):
 img[i,i]=1
for i in range(n):
 img[i,n-1-i]=1
    


plt.figure()
plt.imshow(img)
plt.show()
```
